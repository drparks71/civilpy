## Loads

Input design loads:

```python
# //TODO - Check Civilpy source code for changes to these load labeling conventions
loads = {
    "live": 15.32,
    "dead": 51.22,
    "additional": 5.93,
    "dead_no_fws": 43.72,
}

loads["total_load"] = loads['live'] + loads['dead'] + loads['additional']
loads["total_dead_load"] = loads['dead_no_fws'] + loads['additional']
```

## Bearing Design



```python
from civilpy.structural.aashto.bearings import MethodABearing
```

```python
bearing = MethodABearing(
    width = 7,
    length = 12,
    durometer = 50,
    internal_t = .28,
    external_t = .185,
    steel_t = .105,
    plys = 5,
    span = 12.5,
    expansion_length = 125,
    loads = loads,
    edge_cover = 1/4
)
```
