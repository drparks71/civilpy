````python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# service=Service(ChromeDriverManager().install())
driver = webdriver.Chrome()
url = 'https://www.dot.state.oh.us/Divisions/ContractAdmin/Contracts/BidTabulations/Forms/BidTabs.aspx?RootFolder=%2FDivisions%2FContractAdmin%2FContracts%2FBidTabulations%2FExcel#'
driver.get(url)

time.sleep(2)
````

````python
import codecs

# Get First page
list_of_excel_links = []

elements = driver.find_elements(By.CLASS_NAME, "ms-listlink")

for element in elements:
    text = element.get_attribute('onclick').split(',')
    list_of_excel_links.append("https://www.dot.state.oh.us" + codecs.decode(text[5][1:-21], 'unicode-escape'))
````

````python
import numpy as np
# from civilpy.state.ohio.dot import ohio_counties


def get_county_from_proj_no(proj_no):
    """
    Utilizes digital paper to retrieve the 3 letter county code by searching for that Project Number
    """
    # //TODO - Sometimes the first value doesn't have a county, rewrite to try the second or third if so
    driver.get("https://contracts.dot.state.oh.us/home.do")

    search_box = driver.find_element(By.ID, "DP.PROJECT_NUM.TEXT")
    search_box.send_keys(proj_no)
    go_button = driver.find_element(
        By.XPATH,
        "/html/body/table[3]/tbody/tr[3]/td[2]/table/tbody/tr[1]/td/form/table/tbody/tr[3]/td/table/tbody/tr[2]/td[1]/input[2]",
    )
    go_button.click()

    row_data = driver.find_element(By.CLASS_NAME, "rowdata")

    for column in row_data.text.split("   "):
        if column[:5] == "COUNT":
            county = column.split("  ")[1]
        elif column[:7] == "PID_NUM":
            pid = column.split("  ")[1]
        elif column[:10] == "RT_SECTION":
            cty_rte_sec = column.split("  ")[1]
        elif column[:12] == "PROJECT_TYPE":
            proj_type = column.split("  ")[1]

    # Checks if a county value was found above and assigns NaN if not
    if not "county" in locals():
        county = np.nan

    # Converts 3 Character County Code to Full name
    if county in ohio_counties.values():
        county = list(ohio_counties.keys())[list(ohio_counties.values()).index(county)]

    return county, cty_rte_sec, pid, proj_type
````

## Database layout

3 Tables:

    projects
    contractors
    bid_values

```mermaid
classDiagram
    direction BT
    class bid_values {
       text item_code
       text item_description
       real quantity
       real bid_amount
       integer contractor
       integer project
       integer id
    }
    class contractors {
       text contractor_name
       integer number_of_wins
       integer number_of_second
       integer number_of_third
       integer id
    }
    class projects {
       integer number_of_sections
       text low_bidder
       real low_bid_amount
       text county_cd
       int odot_district
       text item_codes
       int let_date
       integer project_number
    }
    
    bid_values  -->  contractors : contractor
    bid_values  -->  projects : project
    projects  -->  contractors : low_bidder

```

````python
# Create the contractors table
cur.execute(
    """
create table contractors
(
    id               INTEGER not null
        constraint contractors_pk
            primary key autoincrement,
    contractor_name  TEXT,
    number_of_wins   INTEGER,
    number_of_second INTEGER,
    number_of_third  INTEGER
);
"""
)
````

````python
# Create the projects table
cur.execute(
    """
create table projects
(
    project_number     INT not null
        constraint project_number
            primary key,
    number_of_sections INT,
    low_bidder         TEXT
        constraint projects_contractors_id_fk
            references contractors,
    low_bid_amount     REAL,
    county_cd          TEXT,
    odot_district      INT,
    cty_rte_sec        TEXT,
    pid                INT,
    proj_type          TEXT,
    let_date           TEXT,
    client             TEXT
);
"""
)
````

````python
# Create bid_values table
cur.execute(
    """
create table bid_values
(
    id         integer not null
        constraint id
            primary key autoincrement,
    item_code        TEXT,
    item_description TEXT,
    quantity         REAL,
    bid_amount       REAL,
    contractor       TEXT
        constraint contractor_name
            references contractors,
    project    integer not null
        constraint project
            references projects
);
"""
)
````

#### projects Table

````python
def get_projects_from_excel(excel_file):
    """
    Function that takes the ODOT BidTabs Excel File, Loads it into a Dataframe,
    Extracts the necessary details out of it, pushes the formatted table to a database
    and returns the formatted dataframe object

    :param:
        excel_file (Path): Path to the excel file containing the data
    """
    # load the excel file
    excel_book = xlrd.open_workbook(excel_file, encoding_override="iso-8859-1")

    # read by default 1st sheet of an excel file
    df = pd.read_excel(excel_book, header=0)

    # Extract the projects that are included in the data
    projects = pd.DataFrame(df["Project#"].unique(), columns=["project_number"])

    # Create a column for number of sections and fill it with values:
    projects.insert(1, "number_of_sections", None)

    # Fill the above row with the calculated values
    for index, project in projects.iterrows():
        projects.at[index, "number_of_sections"] = len(
            df[df["Project#"] == project.values[0]]["Section"].unique()
        )

    # Create low_bidder column
    projects.insert(2, "low_bidder", None)

    # Fill low_bidder with the values
    for index, project in projects.iterrows():
        projects.at[index, "low_bidder"] = df.loc[
            (df["Project#"] == project.project_number) & (df["BidderOrder"] == 1)
        ]["Name"].values[0]

    # Create low_bid_amount Column
    projects.insert(3, "low_bid_amount", None)

    # Fill low_bid_amount column
    for index, project in projects.iterrows():
        projects.at[index, "low_bid_amount"] = df.loc[
            (df["Project#"] == project.project_number) & (df["BidderOrder"] == 1)
        ]["Bid$"].values[0]

    # Create county_cd Column
    projects.insert(4, "county_cd", None)

    # Fill county_cd column
    for index, project in projects.iterrows():
        try:
            county_code, cty_rte_sec, pid, proj_type = get_county_from_proj_no(
                project.project_number
            )
        except:
            county_code, cty_rte_sec, pid, proj_type = np.nan, np.nan, np.nan, np.nan
        projects.at[index, "county_cd"] = county_code

    # Create odot_distric Column
    projects.insert(5, "odot_district", None)

    # Fill odot_district column
    for index, project in projects.iterrows():
        if project.county_cd in odot_district_by_county:
            projects.at[index, "odot_district"] = odot_district_by_county[
                project.county_cd
            ]
        elif str(project.county_cd)[0] == "D":
            projects.at[index, "odot_district"] = project.county_cd[1:].strip("0")
        else:
            print(project.county_cd)

    # Create cty_rte_sec Column
    projects.insert(6, "cty_rte_sec", None)

    # Fill cty_rte_sec column
    for index, project in projects.iterrows():
        projects.at[index, "cty_rte_sec"] = cty_rte_sec

    # Create pid Column
    projects.insert(7, "pid", None)

    # Fill pid column
    for index, project in projects.iterrows():
        projects.at[index, "pid"] = pid

    # Create proj_type Column
    projects.insert(8, "proj_type", None)

    # Fill proj_type column
    for index, project in projects.iterrows():
        projects.at[index, "proj_type"] = proj_type

    # Create let_date Column
    projects.insert(9, "let_date", None)

    # Store time as string
    for index, project in projects.iterrows():
        excel_date = str(df[df["Project#"] == project.values[0]]["LetDate"].values[0])[
            :10
        ]
        projects.at[index, "let_date"] = excel_date

    # Create client Column - Can use a default value since all projects are ODOT
    projects.insert(10, "client", "ODOT")

    # Push the new table to the database
    projects.to_sql("projects", con, if_exists="append", index=False)
````

#### contractors Table

````python
def get_contractors_from_excel(excel_file):
    """
    Function that takes the ODOT BidTabs Excel File, Loads it into a Dataframe,
    Extracts the necessary details out of it, pushes the formatted table to a database
    and returns the formatted dataframe object

    :param:
        excel_file (Path): Path to the excel file containing the data
    """
    # load the excel file
    excel_book = xlrd.open_workbook(excel_file, encoding_override="iso-8859-1")

    # read by default 1st sheet of an excel file
    df = pd.read_excel(excel_book, header=0)

    # Get the counts for first, second and third place bidders
    contractor_finishes = {}

    # First Place
    for project in list(df["Project#"].unique()):
        contractor = df[(df["Project#"] == project) & (df["BidderOrder"] == 1)][
            "Name"
        ].values[0]

        if contractor in contractor_finishes:
            if 1 in contractor_finishes[contractor]:
                contractor_finishes[contractor][1] += 1
            else:
                contractor_finishes[contractor][1] = 1
        else:
            contractor_finishes[contractor] = {}
            contractor_finishes[contractor][1] = 1

    # Second Place
    for project in list(df["Project#"].unique()):
        try:
            contractor = df[(df["Project#"] == project) & (df["BidderOrder"] == 2)][
                "Name"
            ].values[0]
        except:
            continue

        if contractor in contractor_finishes:
            if 2 in contractor_finishes[contractor]:
                contractor_finishes[contractor][2] += 1
            else:
                contractor_finishes[contractor][2] = 1
        else:
            contractor_finishes[contractor] = {}
            contractor_finishes[contractor][2] = 1

    # Third Place
    for project in list(df["Project#"].unique()):
        try:
            contractor = df[(df["Project#"] == project) & (df["BidderOrder"] == 3)][
                "Name"
            ].values[0]
        except:
            continue

        if contractor in contractor_finishes:
            if 3 in contractor_finishes[contractor]:
                contractor_finishes[contractor][3] += 1
            else:
                contractor_finishes[contractor][3] = 1
        else:
            contractor_finishes[contractor] = {}
            contractor_finishes[contractor][3] = 1

    # Check with the existing entries in the database to avoid duplications
    existing_contractors_values = pd.read_sql_query("SELECT * from contractors", con)
    new_contractors_values = list(pd.DataFrame(df["Name"].unique())[0])

    # Starting with the contractors already in the database, update their values
    for contractor in existing_contractors_values["contractor_name"].values:
        if contractor in contractor_finishes:
            database_index = existing_contractors_values.index[
                existing_contractors_values["contractor_name"] == contractor
            ].values[0]
            current_wins = existing_contractors_values.iloc[database_index][
                "number_of_wins"
            ]
            current_second = existing_contractors_values.iloc[database_index][
                "number_of_second"
            ]
            current_third = existing_contractors_values.iloc[database_index][
                "number_of_third"
            ]

            if 1 in contractor_finishes[contractor]:
                existing_contractors_values.at[database_index, "number_of_wins"] = (
                    current_wins + contractor_finishes[contractor][1]
                )
                del contractor_finishes[contractor][1]
            if 2 in contractor_finishes[contractor]:
                existing_contractors_values.at[database_index, "number_of_second"] = (
                    current_second + contractor_finishes[contractor][2]
                )
                del contractor_finishes[contractor][2]
            if 3 in contractor_finishes[contractor]:
                existing_contractors_values.at[database_index, "number_of_third"] = (
                    current_third + contractor_finishes[contractor][3]
                )
                del contractor_finishes[contractor][3]
            del contractor_finishes[contractor]

    # Add the new contractors to the table
    for contractor in contractor_finishes:
        if 1 in contractor_finishes[contractor]:
            first = contractor_finishes[contractor][1]
            del contractor_finishes[contractor][1]
        else:
            first = 0
        if 2 in contractor_finishes[contractor]:
            second = contractor_finishes[contractor][2]
            del contractor_finishes[contractor][2]
        else:
            second = 0
        if 3 in contractor_finishes[contractor]:
            third = contractor_finishes[contractor][3]
            del contractor_finishes[contractor][3]
        else:
            third = 0

        # Add the new contractors to the dataframe
        existing_contractors_values.loc[len(existing_contractors_values)] = [
            len(existing_contractors_values) + 1,
            contractor,
            first,
            second,
            third,
        ]

    # Push Results to database, this one recreates the table because of the id value
    existing_contractors_values.to_sql(
        "contractors", con, if_exists="replace", index=False, index_label="id"
    )

    return existing_contractors_values
````

#### bid_values Table

````python
def get_bid_values_from_excel(excel_file):
    """
    Function that takes the ODOT BidTabs Excel File, Loads it into a Dataframe,
    Extracts the necessary details out of it, pushes the formatted table to a database
    and returns the formatted dataframe object

    :param:
        excel_file (Path): Path to the excel file containing the data
    """
    # load the excel file
    excel_book = xlrd.open_workbook(excel_file, encoding_override="iso-8859-1")

    # read by default 1st sheet of an excel file
    df = pd.read_excel(excel_book, header=0)

    bid_values = pd.DataFrame()

    # Create item_code Column
    bid_values.insert(0, "item_code", "")
    # Create item_description Column
    bid_values.insert(1, "item_description", "")
    # Create quantity Column
    bid_values.insert(2, "quantity", "")
    # Create bid_amount Column
    bid_values.insert(3, "bid_amount", "")
    # Create contractor Column
    bid_values.insert(4, "contractor", "")
    # Create project Column
    bid_values.insert(5, "project", "")

    # id auto_increments again
    for index, value in df.iterrows():
        contract_query = "SELECT id FROM contractors WHERE contractor_name IS ?"
        try:
            contractor_index = cur.execute(contract_query, (value.Name,)).fetchone()[0]
        except:
            existing_contractors_values = pd.DataFrame(
                {
                    "contractor_name": value.Name,
                    "number_of_wins": 0,
                    "number_of_second": 0,
                    "number_of_third": 0,
                },
                index=[0],
            )
            existing_contractors_values.to_sql(
                "contractors", con, if_exists="append", index=False, index_label="id"
            )
            contractor_index = cur.execute(contract_query, (value.Name,)).fetchone()[0]

        bid_values.at[index, "item_code"] = value["ItemCode"]
        bid_values.at[index, "item_description"] = value["ItemDesc"]
        bid_values.at[index, "quantity"] = value["Qty"]
        bid_values.at[index, "bid_amount"] = value["Bid$"]
        bid_values.at[index, "contractor"] = value["Name"]
        bid_values.at[index, "project"] = value["Project#"]

    # Push Results to database, this one recreates the table because of the id value
    bid_values.to_sql(
        "bid_values", con, if_exists="append", index=False, index_label="id"
    )
````
