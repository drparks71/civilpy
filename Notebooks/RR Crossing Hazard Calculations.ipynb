# Michigan Methodology

Going off [this research paper](https://transportation.org/rail/wp-content/uploads/sites/24/2023/05/Formula-Research-Final-Report.pdf)

Michigan uses a modified version of the New Hampsire Hazard Index

$$ HI = (V) * (T) * (PF) $$

where: 

$HI$ = Calculated Hazard Index  
$V$ = Annual Average Daily Traffic  
$T$ = Train Movements per Day  
$PF$ = Protection Factor:  
- 1.00 - Crossbuck Sign w/ or w/o Yield Sign
- 0.80 - Crossbuck Sign w/ a Stop Sign
- 0.75 - Stop and Flag Procedures
- 0.33 - Flashing Light Signals with any 8" Lenses
- 0.30 - Flashing Light Signals with all 12" Lenses
- 0.27 - Flashing Light Signals with Cantilever Arms and any 8 Lenses
- 0.24 - Flashing Light Signals with Cantilever Arms and all 12 Lenses
- 0.11 - Flashing Light Signals with Roadway Gates and any 8 Lenses
- 0.10 - Flashing Light Signals with Roadway Gates and all 12 Lenses
- 0.09 - Flashing Light Signals with Cantilever Arms and Roadway Gates and any 8 Lenses
- 0.08 - Flashing Light Signals with Cantilever Arms and Roadway Gates and all 12 Lenses
- 0.10 - Any Passive Warning Device with Traffic Signal Interconnection
- 0.10 - Any Active Warning Device with Traffic Signal Interconnection

```python
# Python Formula
def get_michigan_hazard_rating(V: float, T: float, PF: float):
    hazard_index = V * T * PF
    return hazard_index
```
