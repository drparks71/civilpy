### What is Python

Executive summary from [python.org](https://www.python.org/doc/essays/blurb/?external_link=true)

"Python is an interpreted, object-oriented, high-level programming language  
with dynamic semantics. Its high-level built in data structures, combined  
with dynamic typing and dynamic binding, make it very attractive for rapid  
application development, as well as for use as a scripting or glue language  
to connect existing components together. Python's simple, easy to learn  
syntax emphasizes readability and therefore reduces the cost of program  
maintenance. Python supports modules and packages, which encourages program  
modularity and code reuse. The Python interpreter and the extensive  
standard library are available in source or binary form without charge for  
all major platforms, and can be freely distributed."

The aspects of python that benefit engineers are that it has become  
essentially ubiquitous, is included or installable in basically all  
systems, and has essentially accomplished one of the Java programming  
languages original major goals of "write once, run anywhere". Python works  
via an interpreter that users "send" their code to for execution, similar  
to executing VBA code by sending it to excel's VBA intrepreter. The primary  
difference is that VBA operates exclusively within an application such as  
excel, outlook or microstation, while python is generally operated  
externally with the interpreter behaving as it's own stand-alone  
application. The other major difference is that VBA is not open-source, so  
without paying significant sums in licensing fees to Microsoft, we cannot  
include the VBA interpreter itself within a deliverable application sent to  
a client, and have to rely on whatever environment they're running having  
access to the VBA intrepreter. This is generally taken as a given, but when  
we consider the wide range of computing devices, and start including  
devices such as cell phones, tablets, web-servers, IoT devices, sensors,  
bridge control equipment or extreme cases like air-gapped control systems  
for nuclear power plants. Access to Excel or even the internet is no longer  
a given. In general, it is much easier to get python onto those kinds of  
systems than it is to get excel onto them (from a technical standpoint if  
not a bureaucratic one). Anything running any Linux, Mac or essentially any  
of their derivative operating systems like Android or IoS, comes with very  
simple python integration instructions, Windows is also now taking [steps  
to make it easier to add](https://devblogs.microsoft.com/python/python-in-the-windows-10-may-2019-update/). 

The python intrepreter is extremely small and generally consumes very  
small amounts of resources. Due to it being open-source and able to be  
embedded and shipped within software, many software development  
organizations have shed support for VBA in favor for python and "open"  
standards, like those that make up most of the internet such as SSH, HTTPS,  
HTML, CSS, JSON and XML. This trend is expected to continue due to it's  
downstream effect of lowering costs for end users. We have seen python  
interpreters being shipped in applications such as ArcGIS, MicroStation,  
[Autodesk via Revit](https://www.autodesk.com/autodesk-university/class/Getting-Started-Python-2019), Blender, Rhino, AllPlan, Midas and numerous others. 

### What Isn't Python?

Python is not a golden bullet or a solution to all of your problems. It is  
not capable of replicating the life experiences or expertise of a seasoned  
engineer. Although it is frequently used in AI and Machine Learning  
applications, in those instances it heavily relies on "lower level"  
functions and applications written in C/C++. Most of the time users are not  
expected to handle pointers or manage memory to the degree they are in  
lower level languages like C or Rust. This removes a lot of the necessary  
boiler plate code frequently utilized in those languages and significantly  
lowers the barrier to entry for the language, improves readability of the  
code, and preserves the overall beneficial aspects utilizing a programming  
language offers like loops, high levels of customization, error checking,  
testing, and precise control over how an application functions.

### Libraries

Libraries are perhaps the most compelling selling point of the python  
language. Recurring themes of this paper will be **"don't reinvent the  
wheel"** and **"stand on the shoulders of giants"**. There are hundreds of  
incredibly high quality libraries and packages available throughout the  
python ecosystem. The majority of these, following the language's lead, are  
released as free and open-source software. This means there is no cost to  
utilizing or including these packages in your applications.  There are  
numerous libaries for a number of different functions, [matplotlib](https://matplotlib.org/stable/) and  
[plotly](https://plotly.com/) offer users the ability to build out captivating visualizations far  
beyond what's possible in excel or most other off-the-shelf products.  
[Selenium](https://www.selenium.dev/), [requests](https://requests.readthedocs.io/en/latest/), [django](https://www.djangoproject.com/), [flask](https://flask.palletsprojects.com/) and [fastapi](https://fastapi.tiangolo.com/) offer interefaces for you to  
not only make use of just about every resource available on the internet,  
but to build out and deploy new resources yourself. [Handcalcs](https://github.com/connorferster/handcalcs), [latex](https://www.latex-project.org/help/documentation/),  
[jupyter](https://docs.jupyter.org/en/latest/), [pandas](https://pandas.pydata.org/docs/index.html) and [sympy](https://www.sympy.org/en/index.html) give you the ability to perform advanced  
mathematical caculations up to and including advanced calculus and   
differential equations. [Openpyxl](https://openpyxl.readthedocs.io/en/stable/) lets you read and write excel files,  
[pypdf2](https://pypi.org/project/pypdf/) does the same for pdf documents. [Sqlalchemy](https://www.sqlalchemy.org/) provides a generalized

Another net benefit is that code is unforgiving. Although this can  
initially be seen as a negative, it is a positive. Consider this scenario,  
an inexperienced engineer is attempting to solve a problem for which there  
is a built out python package vs. a standardized excel sheet. Utilizing the  
[pint](https://pint.readthedocs.io/en/stable/) library, the developer of the python package made their code unit  
aware, meaning that inputs were given as `4 * units.feet`, rather than just  
`4`. If the user plugged in the incorrect units into the excel sheet, like  
`36` instead of `3` when the intended input was in feet, they would  
generally still recieve a result. If the incorrect units are used as input  
for a python function, the python function will either convert them to the  
necessary unit, or raise an error if it's unable to be converted, python  
will raise an error informing the user that the incorrect inputs were used.

Other aspects of python force you into using additional "Best practices"  
developed by and in use in the computer science industry for years. Things  
like version control utilizing [git](https://www.atlassian.com/git/tutorials/what-is-git), [test driven developemnt](https://testdriven.io/test-driven-development/), and [DRY  
principles](https://www.techtarget.com/whatis/definition/DRY-principle) lend themselves well to developing an application in python,  
but are difficult to properly implement in MathCAD or Excel.

### Other Project Components

Generally a python project consists of not only the actual project files,  
but also all supporting files associated with that project. Examples of  
these files include the projects license, documentation, tests, a readme,  
gitignore files, etc. We'll briefly cover what each of these do, as a  
demonstration, we'll use the [`civilpy`](https://daneparks.com/Dane/civilpy) project. It should be noted that  
these tools are more related to git/version control systems, and as such,  
are not python specific, but are immensely helpful once you start getting  
into managing large projects.

#### [Readme.md](https://daneparks.com/Dane/civilpy/-/blob/master/README.md?ref_type=heads)

Think of this as the "homepage" of a project, at the very least, it should  
contain a short description of what a project does, an installation guide,  
and a couple examples of using the project. There are many tools that can  
be added to a readme.md file that do things like display CI/CD status,  
statistics about the projects, code coverage, etc.

#### [.gitignore](https://daneparks.com/Dane/civilpy/-/blob/master/.gitignore?ref_type=heads)

This is part of the git version controlling system. Files you do **NOT**  
want tracked are listed here. Examples of this would include things like  
outputs of an application, secrets like passwords or API keys, or raw data  
your application reads that is constantly being updated and is hosted on a  
different source. Including these types of files is either unnecessary or  
dangerous in the case of API keys or secrets. Often projects require users  
to create a file called `secrets.txt` or `secrets.json` in the root  
directory. The application can then import the necessary key/value from  
those files and use it to ensure only the right people have access to the  
necessary systems.

#### [CI/CD Yaml Files](https://daneparks.com/Dane/civilpy/-/blob/master/.gitlab-ci.yml?ref_type=heads)

These are commonly included in python projects, [here's a brief description  
of CI/CD](https://about.gitlab.com/topics/ci-cd/). Basically, sometimes you want automated functions to be run when  
you update your project, things like tests, building out the website/UI  
your project controls, sending notifications, deploying your project out to  
various sources etc. 

#### [Docs Folder and .readthedocs.yaml](https://daneparks.com/Dane/civilpy/-/blob/master/.readthedocs.yaml?ref_type=heads)

Documentation is one of, if not the most important part of your project. It  
explains in detail how a project functions. There are many automated tools  
to assist users in building out their documentation like [sphinx](https://docs.readthedocs.io/en/stable/intro/getting-started-with-sphinx.html). At the  
very least, any function defined in your source code should have something  
called a docstring. Docstrings are included in python code by using triple  
quotes syntax that results in multiline comments, consider the following  
example;

```python
def find_key_for_value_in_tuple_range(data, number):
    """
    Find the key in a dictionary where the number falls between the first 
    two values of the tuple.

    Args:
        data (dict): Dictionary containing lists of tuples.
        number (float): The number to search for within the tuple ranges.

    Returns:
        key: The key in the dictionary whose tuple contains the number, or None 
            if not found.
    """
    for key, values in data.items():
        if len(values) >= 2 and values[0].magnitude <= number <= values[1].magnitude:
            return key
    return None
```

In the above function, the docustring is longer than the actual function  
being defined. These docstrings are what sphinx uses to build out  
documentation. There's multiple format styles available within sphinx, the  
above is an example of Google's standard. More on docstrings can be found  
[here.](https://www.dataquest.io/blog/documenting-in-python-with-docstrings/#:~:text=Docstrings%20are%20for%20using%20the,two%20are%20the%20most%20common.)

Example of projects with good documentation are [scikit-learn](https://scikit-learn.org/stable/) and [pandas](https://pandas.pydata.org/docs/index.html).

When writing docstrings at Michael Baker, the [Google style docstring](https://sphinxcontrib-napoleon.readthedocs.io/en/latest/example_google.html) is  
the preferred style.

#### [License](https://daneparks.com/Dane/civilpy/-/blob/master/LICENSE?ref_type=heads)

Licensing can get very complicated. Escalate issues whenever you're unsure  
if you're able to use a specific package/license.

## Notebook Formatting and Structure

Notebooks should be formatted in a way that maintains their ability to stay  
readable and compliant with [PEP 8](https://peps.python.org/pep-0008/). PEP 8 only covers python code itself  
though, and not notebooks. So it should be seen as the fall back guidance  
with this document as having precedence for structuring concerns.

Style guides defined in PEP8, like spaces over tabs and 4 spaces to an  
indent are Baker's preferred style guidelines unless overriden within this  
document. This should ensure that the default python settings for things  
like IDEs and Black just naturally enforce Baker's preferences.

Users would benefit from implementing the following rulers in their jupyter  
environments and other IDEs,

ruler settings can be found under the "settings"->"Settings Editor", or by  
typing <kbd>ctl</kbd> + <kbd>,</kbd>. Within the settings editor select "Notebook", and scroll  
down to "Code Cell Configuration" and "Markdown Cell Configuration" to  
adjust these values. Note that these aren't bold in the settings menu,  
and can be easy to miss.

for "Code Cells" setting a ruler at 79 helps ensure compliance with  
python's PEP8,



in addition this ensure code will not be "cut short" when outputting  
notebooks as pdfs when generating deliverables.

Markdown cells should have a ruler set at 75. Rather than allowing the text  
to wrap each line should end with a double space so the text presented to  
the reader is less than 85 characters long. It should be noted that this  
rule can be ignored for characters not displayed to the user, for instance,  
the link `[click]('https://google.com')` should be regarded as 5 characters  
for the purpose of character counts. This is because the 75 character  
guidance is based on [this research](https://baymard.com/blog/line-length-readability), which has found that digital text is  
most readable to users when presented between 50 and 75 characters at a  
time. This also prevents issues when rendering the text as a pdf as the  
line formatting will appear as the user intended.



Generally notebooks do not handle large table outputs from pandas well.  
This is something engineers should consider when utilizing python for calc  
packages. An alternative method for longer outputs might be a process like  
outputting the table to excel, and then adjusting the page layout there  
before outputting the results as a pdf.

## What Can Python Be Used For

There's not a lot that python, like any turing-complete language, can't  
accomplish. A better question is "What **should** python be used for?".  
Going back to the second theme of this paper, **don't reinvent the wheel**.  
Although a python application could be built out to perform FEA, there's  
already an application readily available to all MB employees to do that via  
MIDAS. A better use of an employees time, would be to develop the  
customizations necessary for frequently performed tasks in MIDAS,  
utilizing the MIDAS API. Basic simple "verification calcs" are easy enough  
to write out in python and then be run on various documents. 

More advanced topics would be things like developing QA/QC tools that can  
be given a pdf document and check certain things about it. For instance,  
that any of the markups that have an attribute related to their current  
status have all be toggled to "closed" indicating that all of the comments  
have been addressed. This can speed up QA processes and help focus the  
quality auditors time into areas that have a higher potential to contain  
errors.

The most complicated topics, include things like building full blown user  
interfaces and things like developing customized machine learning  
applications trained on MB data. This is where the development effort  
starts to get beyond a sole developer and starts to get more into a team  
based effort.

In general it's better for code to be kept in a centralized location like  
[github](https://github.com/) where the discussion and community effort required to go into  
building advanced code bases can be stored alongside the code itself. The  
intricacies of a version control system like git are beyond the scope of  
this document, but more about it can be read [here](https://learn.microsoft.com/en-us/training/modules/intro-to-git/). One of the best ways to  
reduce duplicative efforts, is to have an obvious place for users to look  
first before engaging in that effort. If their same problem has already  
been solved, this can save them significant amounts of time. If a similar  
problem has been solved, they can help to expand that code base to address  
the additional features they need.

### Outputting Notebooks to PDF Format for Deliverables

````python
import asyncio
import nbformat
from nbconvert import WebPDFExporter
from nbconvert.preprocessors import TagRemovePreprocessor


def notebook_to_pdf(notebook_path):
    # Set the appropriate event loop policy for Windows
    if asyncio.get_event_loop().is_running():
        asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())

    # Read the notebook
    with open(notebook_path) as f:
        notebook = nbformat.read(f, as_version=4)

    # Configure the tag removal preprocessor
    tag_remove_preprocessor = TagRemovePreprocessor()
    tag_remove_preprocessor.remove_cell_tags = ('remove_cell',)
    tag_remove_preprocessor.remove_single_output_tags = ('remove_output',)
    tag_remove_preprocessor.remove_input_tags = ('remove_input',)

    # Create the WebPDF exporter and register the preprocessor
    pdf_exporter = WebPDFExporter()
    pdf_exporter.register_preprocessor(tag_remove_preprocessor, enabled=True)

    # Convert the notebook to PDF
    pdf_data, resources = pdf_exporter.from_notebook_node(notebook)

    # Save the PDF to a file
    pdf_filename = notebook_path.replace(".ipynb", ".pdf")
    with open(pdf_filename, "wb") as f:
        f.write(pdf_data)
    print(f"PDF created: {pdf_filename}")
````
