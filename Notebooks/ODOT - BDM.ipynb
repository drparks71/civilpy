### 101.2 - BDM Audience

The requirements of the two-column format ODOT Bridge Design Manual (BDM)  
are written to the "Designer of Record" as defined in BDM Section 101.4.  
The sentences that direct the Designer of Record to perform work are  
written as commands. For example, a requirement to design bridges for HL-93  
loading would be expressed as, "Design all bridges for an HL-93 loading,"  
rather than "The Designer of Record shall design all bridges for an HL-93  
loading." In the imperative mood, the subject, "the Designer of Record" is  
understood.

All requirements to be performed by others have been written in the active  
voice. Sentences written in the active voice identify the party responsible  
for performing the action. For example, "The Office of Structural  
Engineering will review designs for non-standard bridge railings."

Sentences that define terms; describe a product or desired result; or  
describe a condition that may exist are written in indicative mood. These  
types of sentences use verbs requiring no action. For example, "All  
reinforcing steel shall be epoxy coated."

### 101.3 - Purpose

The BDM establishes ODOT policy for the following:

1. The analysis, design and rating of new and existing bridges.
2. The analysis and design of new and existing retaining walls and noise  
   walls.
3. The analysis and design of new and existing pedestrian structures.

Consider the practicability of construction with reference to each detail  
of design.

### 101.7 - Definitions

<u>**Bridge**</u>  
The ORC 5501.47(B)(1)(c) defines a bridge as any structure of ten feet or  
more clear span or ten feet or more in diameter on, above, or below a  
highway. Refer to the Ohio Bridge Inventory Coding Guide for more specific  
information.

<u>**Bridge Limits**</u>  
The distance between the bridge ends of the approach slabs measured to the  
nearest hundredth of a foot along the Centerline of Construction.

<u>**Centerline of Construction**</u>  
The reference line used for construction of a project. Normally located at  
the median centerline on a divided highway or at the normal crown point  
location on an undivided highway. 

<u>**Checker**</u>  
The Checker is an individual employed by the Design Agency responsible for  
ensuring correctness, constructability and completeness of designs detailed  
in the Contract Documents.

<u>**Contract Documents**</u>  
Refer to definition provided in C&MS 101.03.

<u>**Culvert**</u>  
Unless otherwise noted, a culvert shall be defined herein as a buried  
structure type, regardless of span, placed on, above or below a highway.

<u>**Design Agency**</u>  
The firm, partnership, association, limited liability company, government  
entity or corporation that employs the Designer of Record in accordance  
with ORC 4733.

<u>**Design Loading**</u>  
The envelope of load factors and load combinations encompassing the  
permanent and transient loads acting on the structure as required by the  
governing design specifications.

<u>**Designer**</u>  
The Designer is an individual employed by the Design Agency responsible for  
preparing designs detailed in the Contract Documents.

<u>**Designer of Record**</u>  
The Designer of Record is the individual that professionally endorses  
(signs and seals) the design in the Contract Documents in accordance with  
OAC 4733-23-01(C) and as specified in the L&D, Vol. 3 Section 1302.6.1. 

<u>**Haunch**</u>  
The distance between the top of a beam/girder and the bottom of the deck.

<u>**Invert**</u>  
The bottom or lowest point of the internal surface of the transverse cross  
section of a pipe or conduit 

<u>**Normal Water Elevation**</u>  
Water elevation within a waterway produced by groundwater flow and not  
influenced or minimally influenced by surface water runoff. The Normal  
Water Elevation shown in the plans is approximate and will fluctuate  
seasonally and from year to year.

<u>**Ordinary High Water Mark**</u>  
Defined by the USACE Regulatory Letter No. 05-05 which is available [here.](https://www.usace.army.mil/Missions/CivilWorks/RegulatoryProgramandPermits/GuidanceLetters.aspx)  
This elevation marks the jurisdictional limit of the USACE.

<u>**Owner**</u>  
The Owner shall be defined as the duly authorized agent of the government  
agency sponsoring the project. For ODOT sponsored projects, the sponsoring  
District shall be considered the Owner.

<u>**Pedestrian Structure**</u>  
A structure which is designed for and intended to primarily carry  
pedestrians, bicyclists, equestrian riders, and light maintenance vehicles,  
but not designed for and intended to carry typical highway traffic. 

<u>**Proposal Notes**</u>  
Specifications that supplement and supersede Supplemental Specifications  
and the C&MS. 

<u>**Project Manager**</u>  
Individual selected by the sponsoring agency responsible for advancing the  
project through the development process. Refer to ODOT’s Project  
Development Process Manual for specific duties.

<u>**Reference Line**</u>  
A chord drawn from centerline to centerline of abutment bearings at the  
Centerline of Construction or an extension of a line tangent to the  
Centerline of Construction at the centerline of abutment bearings.

<u>**Rehabilitation**</u>  
Title 23 Part 650 Section 403 of the Code of Federal Regulations (i.e. 23  
CFR 650.403) defines "Rehabilitation" as the major work required to restore  
the structural integrity of a bridge as well as work necessary to correct  
major safety defects. Examples include: deck replacement, fatigue retrofit,  
structural patching, bearing replacement and railing upgrade.

<u>**Reviewer**</u>  
The Reviewer is an individual employed by the Design Agency responsible for  
the overall evaluation of the design in the Contract Documents for  
completeness, consistency, continuity, constructability, general design  
logic and quality.

<u>**Skew angle**</u>  
The skew angle is the angle of deviation of the substructure unit from  
perpendicular to the centerline of construction or reference chord. 

<u>**Supplemental Specifications**</u>  
Specifications that supplement and supersede the C&MS. 

<u>**Supplements**</u>  
Individually numbered documents describing necessary information such as  
laboratory test methods and certification or pre-qualification procedures  
for materials.

<u>**Temporary Access Fill**</u>  
A fill or structure that allows a contractor access to work on roads or  
bridges located within bodies of water. Examples of TAF’s include:  
cofferdams; temporary structures for maintaining traffic; causeways and  
work pads; and demolition debris.

<u>**Thalweg**</u>  
The line defining the points of lowest elevation along the length of a  
riverbed. Also, commonly referred to as the Flow Line.

Provide divided highways having four or more lanes crossing under an

<div class="alert alert-block alert-info">

### 302.4 - Interference Due to Exisiting Substructure

Do not locate a new substructure unit where an existing substructure or its  
foundation would interfere with the foundation or construction of the  
proposed substructure unit. 

<div class="alert alert-block alert-info">
<b>C302.4:</b><br>
    This applies particularly where piles are to be driven.  It is desirable to<br> 
    avoid the difficulty and expense of removing deep underground portions of<br> 
    the existing substructure and to avoid the resultant disturbance of the<br> 
    ground. Obstructions within the footprint of the new substructure shall be<br> 
    completely removed to avoid interference.  Where existing substructure<br> 
    units are shown on the Site Plan, place the locations precisely. Show the<br> 
    existing substructure configuration based on existing plans or field<br> 
    verified dimensions. Misrepresentation of the location of the existing<br>
    substructure units has resulted in expensive change orders during<br> 
    construction. 
</div>

Label existing dimensions as (+/-) plus or minus in the contract documents. 

### 302.5 - Bridge Structure, Skew, Curvature and Superelevation

During preliminary engineering, investigate alternatives that eliminate the  
presence of skew angles that exceed 30°, horizontal curves and  
superelevation transitions within the actual bridge limits. 

<div class="alert alert-block alert-info">
<b>C302.5:</b><br>
    Skews, curves and superelevation transitions add complexity which can<br>  
    increase costs and also result in poor ride quality. 
</div>

## 303 - Design Loading

### 303.1 - Highway Bridges

#### 303.1.1 - Vehicular Live Load

Design all bridges that carry highway traffic for an HL-93 loading as  
specified in LRFD 3.6.1.2.1.

#### 303.1.2 - Future Wearing Surface

Design all bridges that carry highway traffic for a future wearing surface  
(FWS) of 0.060-ksf except as noted in BDM Section 401.3.

#### 303.1.3 - Longitudinal Forces

Assume force effects transferred through bearings to the substructure occur  
at the bearing seat elevation. 

<div class="alert alert-block alert-info">
<b>C303.1.3:</b><br>
    Bearings that permit rotation (e.g. elastomeric bearings, rocker bearings,<br>
    etc.) will develop eccentricities resulting in applied moments as defined<br>
    in LRFD 14.6.3.2. Typically, these eccentricities are negligible with<br>
    respect to the design of the substructures that support the bearings and<br>
    may be ignored. 
</div>

The total factored longitudinal loading applied to the substructure at each  
expansion bearing shall be the larger of: 

1. The Frictional Force acting on the substructure if the bearing slides.  
1. The bearing’s nominal (i.e. unfactored) resistance to longitudinal  
loading. 

<div class="alert alert-block alert-info">
<b>C303.1.3:</b><br>
    A properly designed expansion bearing will utilize friction in lieu of<br>
    mechanical anchorage to prevent the bearing from sliding under lateral <br>
    loading. For these bearings, the maximum lateral loading on the <br>
    substructure unit will be in the form of Friction, LRFD 3.13.<br><br>    
    For bearings that require mechanical anchorage, the bearing’s resistance to<br>  
    lateral loading represents the maximum lateral load on the substructure <br>
    unit. Resistance in this instance is nominal because it is applied to the <br>
    substructure as a loading.<br>
</div>

#### 303.1.4 - Seismic

All bridges in the State of Ohio are located within Seismic Performance  
Zone 1. 

<div class="alert alert-block alert-info">
<b>C303.1.4:</b><br>
    This provision is based on Ohio’s seismologic data which is explained in<br>
    BDM Section 305.1.5. 
</div>

Refer to BDM Section 305.1.5 for Seismic Geotechnical requirements. 

##### 303.1.4.1 - Seismic Performance Zone 1

Seismic analysis is not required except as noted in BDM Sections 303.1.4.1.a  
and 303.1.4.1.b.

###### 303.1.4.1.b - Horizontal Connection Force

All structures shall have some mechanism to transfer horizontally applied  
superstructure loads (e.g. vehicular braking force, centrifugal force,  
vehicular collision force, friction load, water load, wind load, and wind  
load on live load) to the substructure to ensure structural stability.  
During a seismic event, design these mechanisms, that prevent the free  
horizontal translation of the superstructure in any direction relative to  
the substructure, to transfer an applied horizontal connection force at the  
Extreme Event I Limit State. Only provide additional restraint for seismic  
loads (e.g. seismic pedestals) where the mechanisms noted above do not  
provide sufficient capacity. Refer to BDM Section 303.1.4.1.c for bearing  
requirements. 

<div class="alert alert-block alert-info">
<b>303.1.4.1.b:</b><br>
    Examples of mechanisms include fixed bearings, bearing guides, abutment<br>
    diaphragms, diaphragm guides, wing walls and wind locks. 
</div>

The magnitude of the connection force shall be 0.15 or 0.25 times the  
tributary permanent load at the location of the restraint as determined in  
LRFD 3.10.9.2. If sufficient geotechnical information is not available to  
determine Site Class, assume the magnitude of the connection force is 0.25  
times the tributary permanent load. 

<div class="alert alert-block alert-info">
<b>303.1.4.1.b:</b><br>
    The load factor for live load, γEQ, may be taken as 0.0.<br><br>
    The tributary permanent load defined in LRFD 3.10.9.2 represents the<br>
    factored dead load of the superstructure applying load to the device or<br>
    object providing the directional restraint. If every bearing supporting the<br>
    superstructure provides transverse restraint, the tributary permanent load<br>
    applied to each restraint would equal the factored dead load reaction at<br>
    each bearing. If only one transverse restraint was provided at each<br>
    substructure unit, the tributary permanent load applied to each restraint<br>
    would equal the sum of the factored dead load reactions for each bearing at<br>
    the substructure unit. If only one transverse restraint was provided for<br>
    the entire superstructure unit, the tributary permanent load applied to the<br>
    restraint would equal the sum of the factored dead load reactions of every<br>
    bearing. Longitudinal restraint connection forces would be determined<br>
    similarly. For restraint provided in multiple directions, LRFD 3.10.8<br>
    applies.
</div>
