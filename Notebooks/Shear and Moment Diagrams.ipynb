```python
from civilpy.structural import beam_analysis
```

```python
L = 5

beam = beam_analysis.newEulerBeam(L)
```

```python
# Define beam and fixities
pinned = [1,1,0]
roller = [0,1,0]
beam.setFixity(0, pinned, label = 'F')
beam.setFixity(L, roller, label = 'E')

# Define loads
Pz = -1
beam.addVerticalLoad(0, Pz, label = 'A')
beam.addVerticalLoad(L*0.5, 2*Pz, label = 'B')
beam.addVerticalLoad(L, Pz, label = 'C')
beam.addDistLoadVertical(0, L, Pz)
beam.addDistLoadVertical(1, L*0.3, 5*Pz)

# Plot the beam diagram
beam_analysis.plotBeamDiagram(beam)

# Run the analysis
analysis = beam_analysis.PyNiteAnalyzer2D(beam)
analysis.runAnalysis()

# Plot the SFD and BMD
beam_analysis.plotShear(beam)
beam_analysis.plotMoment(beam)
```
