[Jupyterlite](https://jupyter.org/try-jupyter/lab/) - WASM based, early in it's development but promising, requires no installation and runs directly from the browser. Only allows the use of pure python libraries and can't interact with local files without significant amounts of additional work.

## Setting up Python

If you don't have python, download it, either from [here](https://www.python.org/downloads/) or from the windows store. Which version of python to use is a difficult question to answer. Generally you want to refer to [this diagram](https://devguide.python.org/versions/) and select a version that's not so new that it still has a lot of issues, but not so old that it's no longer supported. As of writing (5/5/2024) I'd recommend users download [python 3.10.14](https://www.python.org/downloads/release/python-31014/) which is the most updated version of the 3.10 versions. Note that if you want to use a specific library, and it was written when a different version of python was available, you'll likely have to download that version of python to be able to use it. Python 3.8 generally provides good stability and access to most of python's modern features, but is rapidly approaching end-of-life as far as support as of writing. If you come accross a library that utilizes python 2.X, my advice is to find another way to accomplish your goals.

### Configuring Your kernel

In order to use the python you just downloaded within Azure Data Studio. You have to point the kernel to the version of python you just installed. Azure data studio should normally be able to find the location of your python interpreter, but if it can't you can open a terminal window, type `python` and then these three commands in order,

```python
import os`
import sys
os.path.dirname(sys.executable)
```

which will return the location of you python executable. In azure data studio, either open one of your recently downloaded notebooks, or create a new notebook.

The first time you configure a python interpreter as your kernel,



Azure data studio should provide you a wizard to select the version of python you want, or to install a new one. Python relies on a system called `virtual environments` where each project generally has it's own environment, so if you want to install a new version of python from here, that's fine too. [Here's microsoft's documentation](https://learn.microsoft.com/en-us/azure-data-studio/notebooks/notebooks-python-kernel) Which provides similar instructions to what I'm providign here.

If the wizard doesn't work or you want to manually configure which version of python to use, go to `File`->`Preferences`->`Settings`



search python in the settings,



and enter the path to your python installation that you found earlier with the `os.path.dirname(sys.executable)` command. Check the box that says `Local Path to a prexisting python...` and begin installing python packages.

# Anaconda

Anaconda is my preferred method of running notebooks. They define the standard and most of the features seem to produce the most expected results when run in their software.

[Download and install anaconda](https://docs.anaconda.com/free/anaconda/install/index.html)

Run anaconda prompt,



## Create a New Environment 

by running `conda create -n civilpy python=3.10` replace civilpy with whatever you want to name your environment, and 3.10 with whichever version of python you want to use.

## Install packages

Activate the environment you just created with the command `conda activate civilpy` then install the required packages with `pip install civilpy`

once it's done running the installation script, run `jupyter lab`

if you want to run or edit the notebooks from the civilpy package, first navigate to wherever you downloaded the Notebooks folder with `cd "C:\Path\to\civilpy\Notebooks"` then run `jupyter lab`.

For all future times when you want to run jupyter you no longer have to create or install the packages, just run;

```bash
(base) $ conda activate civilpy
(civilpy $ jupyter lab
```

to stop the application, just close your browser window and terminal in which it's running.

# Pycharm

[Install Pycharm](https://www.jetbrains.com/pycharm/)

Note that jupyter notebooks aren't able to run natively in the free version of pycharm. However, you can open a terminal after configuring the jupyter environment and run `jupyter lab` just like with you conda environment which will open jupyter in the browser for you.
