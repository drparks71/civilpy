```python
import restapi
import os
import time
from pathlib import Path

full_run_time_start = time.time() 

url = 'https://services3.arcgis.com/6LvtIYUSMXW8Tb6o/ArcGIS/rest/services'

# reference map service
ags = restapi.ArcServer(url)
```

```python
from pprint import pprint

for root, services in ags.walk():
     pprint('Folder: {}'.format(root))
     print(f'Services: ')
     [print(service) for service in services]
```

```python
metra_bridges = ags.getService('Metra_Bridges/FeatureServer')
```

```python
metra_bridges
```

```python
for service in metra_bridges:
     print('name: "{}"'.format(service.name))
     print('repr: "{}"'.format(repr(service)))
```

```python
where = '1=1'
```

```python
bridges = metra_bridges.layer(0)
```

```python
featureSet = bridges.query(exceed_limit=True)
```

```python
len(featureSet.json.features)
```

```python
import folium
import geopandas as gpd

# Create a Map instance
m = folium.Map(location=[41.8781, -87.6298], zoom_start=10, control_scale=True)

# Add json data to map
folium.GeoJson(
    featureSet.json,
    name="Metra Bridges",
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(icon='star')),
    tooltip=folium.GeoJsonTooltip(fields=["BRNO", "CROSSING", "LINE"]),
    popup=folium.GeoJsonPopup(fields=["BRNO", "CROSSING", "FULLCODE", "CROSSING", "CITY", "LINE", "LINE_1", "MILE", "NOTES", "PLAN_17"]),
).add_to(m)
```

```python
crossings = fra_crossings.layer(0)

featureSet = crossings.query(exceed_limit=True)
```

```python
m = folium.Map(location=[41.8781, -87.6298], zoom_start=10, control_scale=True)

# Add json data to map
folium.GeoJson(
    featureSet.json,
    name="FRA Crossings",
    zoom_on_click=True,
    marker=folium.Marker(icon=folium.Icon(icon='star')),
    tooltip=folium.GeoJsonTooltip(fields=["CROSSING", "RAILROAD", "STREET"]),
    popup=folium.GeoJsonPopup(fields=["CROSSING", "RAILROAD", "STREET", "CROSSING", "CITYCD", "ACC_LINK", "INV_LINK", "RAILROAD", "RRDIV", "RRSUBDIV"]),
).add_to(m)
```
