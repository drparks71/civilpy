```python
# Pandas doesn't read hyperlinks - would be better to have the explicit address in the excel book
import pandas as pd
# Load the excel sheet, 5th tab, Column 'R'=index 17
df = pd.read_excel(path_to_excel, sheet_name=4)
```

```python
# Store Column in list
mylist = df['Unnamed: 17'].tolist()
```

```python
# Filter the empty rows from the list
filtered_list = [(index, value) for index, value in enumerate(mylist) if len(str(value))>14]
```

```python
# Hacky workaround for pandas not being able to read urls, loads whole wb,
# So it's pretty slow, read-only is faster, but links break
import openpyxl

wb = openpyxl.load_workbook(path_to_excel)
ws = wb['New Bridges']
```

```python
url_list = []
sfn_list = []

# This will fail if there is no hyperlink to target
for i, _ in filtered_list:
    url_list.append(ws.cell(row=i+2, column=18).hyperlink.target)
for i, _ in filtered_list:
    sfn_list.append(ws.cell(row=i+2, column=1).value)
```

```python
def find_first_iframe(driver):
    iframes = driver.find_elements_by_xpath("//iframe")
    for index, iframe in enumerate(iframes):
        # Your sweet business logic applied to iframe goes here.
        driver.switch_to.frame(index)
        try:
            view_pdf_1 = driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_upRASContent"]/table/tbody/tr[3]/td[2]/span')
            view_pdf_1.click()
        except:
            pass

        find_first_iframe(driver)
        driver.switch_to.parent_frame()
```

```python
def find_second_iframe(driver):
    iframes = driver.find_elements_by_xpath("//iframe")
    for index, iframe in enumerate(iframes):
        # Your sweet business logic applied to iframe goes here.
        driver.switch_to.frame(index)
        try:
            view_pdf_2 = driver.find_element(By.XPATH, '//*[@id="ContentPlaceHolder1_printSections_cmdViewPDFBottom"]')
            view_pdf_2.click()
        except:
            pass

        find_second_iframe(driver)
        driver.switch_to.parent_frame()
```

```python
# Loop through the list opening the link in selenium
# Has to be full screen
wait = WebDriverWait(driver, timeout=20)

for url, sfn in zip(url_list, sfn_list):
    print(url)
    driver.get(url)
    wait_1 = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="tabAsset"]/span/span/span')))
    asset_info = driver.find_element(By.XPATH, '//*[@id="tabAsset"]/span/span/span')
    asset_info.click()
    
    wait_2 = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="completedReportsGridDiv"]/div/div[2]/div/div[1]/div[3]/div[1]/div/div[10]/div')))
    hamburger = driver.find_element(By.XPATH, '//*[@id="completedReportsGridDiv"]/div/div[2]/div/div[1]/div[3]/div[1]/div/div[10]/div')
    hamburger.click()

    # finds the button in the first iframe
    find_first_iframe(driver)
    
    # finds the button in the second iframe
    find_second_iframe(driver)
```

```python
def rename_files():
    for index, sfn in enumerate(sfn_list):
        if index == 0:
            # Renames the file
            try:
                shutil.copy(Path(file_save_path) / "report.pdf", Path(file_save_path) / f"{sfn}_inspection_report.pdf")
                os.remove(Path(file_save_path) / "report.pdf")
                time.sleep(5)
            # If source and destination are same
            except shutil.SameFileError:
                print("Source and destination represents the same file.")

            # If there is any permission issue
            except PermissionError:
                print("Permission denied.")

            # For other errors
            except:
                print("Error occurred while copying file.")
        else:
            # Renames the file
            try:
                shutil.copy(Path(file_save_path) / f"report ({index}).pdf", Path(file_save_path) / f"{sfn}_inspection_report.pdf")
                os.remove(Path(file_save_path) / f"report ({index}).pdf")
                time.sleep(5)
            # If source and destination are same
            except shutil.SameFileError:
                print("Source and destination represents the same file.")

            # If there is any permission issue
            except PermissionError:
                print("Permission denied.")

            # For other errors
            except Exception as e:
                print(f"Error occurred while copying file. {e}")
```

```python
rename_files()
```
