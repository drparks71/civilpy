# NLTK in Python

[NLTK](https://www.nltk.org/) - Natural Language Toolkit is a series of python tools created to aid in the extraction, manipulation and analysis of data regarding [information stored in english language structures](https://realpython.com/nltk-nlp-python/#tokenizing). This relates to civil engineering mostly in regards to plans. Through extracting and cleaning the words used within plan sets, we can more easily optimize indexing various plan sets so that search engines are able to more accurately return results relevant to a given query (think Google vs. Windows search). This is not getting into machine learning and AI yet, these are still the foundational steps that are required in order to have sufficient data for eventually training AI models.

## NLTK Basics

Tokenizing: The process of splitting a sentence into it's constituent parts and labeling them.

````python
import nltk
````

````python
sentence = "At eight o'clock on Thursday Morning Arthur didn't feel very good."
````

````python
tokens = nltk.word_tokenize(sentence)
tokens
````

````python
tagged_words = nltk.pos_tag(tokens)
tagged_words
````

To get a list of what the tags mean, run the following command, they all represent parts of speach
```python
nltk.help.upenn_tagset()
```

In addition to tokenizing by words, you can tokenize by sentences,

````python
from nltk.tokenize import sent_tokenize
````

````python
example_string = """I must not fear. Fear is the mind-killer. Fear is the little-death that brings total obliteration. I will face my fear. I will permit it to pass over me and through me. And when it has gone past I will turn the inner eye to see its path. Where the fear has gone there will be nothing. Only I will remain."""
````

````python
sent_tokenize(example_string)
````
